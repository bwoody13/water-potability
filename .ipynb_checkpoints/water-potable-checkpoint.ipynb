## Training Models
Some things I want to keep in mind during the training of thses models is that we want to not only have as good classification accuracy as possible, but also ensure we reduce the number of false positives in our predictions. We want to try and avoid these since it would be much worse to say that non-potable water is potable than the other way around since many people could get sick if a model that did that was in production.
*Note: Whenever the use of the Imputer is mentioned, it means the data with N/A values was kept and all N/As were replaced with the mean. For standard models we use the data that removed N/As in the beginning.*

#### Helper Functions
Some functions used throughout the models to make the code simpler

##### Scoring

```python
def predict_and_score(model, X, y, **kwargs):
    y_pred = model.predict(X)
    return score(y, y_pred, **kwargs)

def score(y,  y_pred, output='all', additional_print="", ret_data=False):
    acc = accuracy_score(y, y_pred)
    conf_mat = confusion_matrix(y, y_pred)
    clf_rep = classification_report(y, y_pred)
    acc_print = f"{additional_print} Accuracy={acc}"
    if output == 'all':
        print(acc_print)
        print("Confusion Matrix:", f"Num FP={conf_mat[0, 1]}\n", conf_mat)
        print("Classification Report:\n", clf_rep)
    elif output == 'acc':
        print(acc_print)
    if ret_data:
        return acc, conf_mat, clf_rep
    
def multi_score(model, X, y, runs=10, progress=False, output='all', additional_print="", ret_data=False):
    total_acc = 0
    for i in range(runs):
        ps_output = 'acc' if progress else None
        acc, conf_mat, clf_rep = predict_and_score(model, X, y, output=ps_output, ret_data=True)
        total_acc += acc
    
    avg_acc = total_acc / runs
    acc_print = f"For {runs} runs: Avg {additional_print} Accuracy={acc}"
    if output == 'all':
        print(acc_print)
        print("Confusion Matrix:", f"Num FP={conf_mat[0, 1]}\n", conf_mat)
        print("Classification Report:\n", clf_rep)
    elif output == 'acc':
        print(acc_print)
    if ret_data:
        return acc, conf_mat, clf_rep
    
```

##### Fitting

```python
def transform_clf_params(params):
    return {f"classifier__{key}": val for key, val in params.items()} 

scorers = {
    'll': make_scorer(log_loss, greater_is_better=False, needs_proba=True),
    'mcc': make_scorer(matthews_corrcoef),
    'ps': make_scorer(precision_score),
    'acc': make_scorer(accuracy_score),
    'auc_pr': make_scorer(average_precision_score),
    'roc_auc': make_scorer(roc_auc_score)
}

def grid_search_fit(model, X, y, params={'C': [0.001, 0.01, 0.1, 1, 10]}, cv=10, output=True, confusion=True, class_rep=True, is_pipeline=False, refit='acc'):
    if is_pipeline:
        params = transform_clf_params(params)
    grid_search = GridSearchCV(model, params, cv=cv, scoring=scorers, refit=refit)
    grid_search.fit(X, y)
    best_params = grid_search.best_params_
    best_score = grid_search.best_score_
    best_model = grid_search.best_estimator_
    if output:
        print(f"Grid Search with CV found best params={best_params}, best score={best_score}")
    return best_model
```

##### Pipeline

```python
def make_pipeline(model, imputer=False, scaler=True, selector=False, selector_to_use='kbest', n_features=7, poly_feats=False, poly_degree=(2,3)):
    pipeline = []
    
    if imputer:
        pipeline.append(('imputer', SimpleImputer(strategy='mean')))
    
    if poly_feats:
        PolynomialFeatures(degree=poly_degree)
    
    if scaler:
        pipeline.append(('scaler', StandardScaler()))
    
    if selector:
        selectors = {'kbest': SelectKBest(score_func=mutual_info_classif, k=n_features),
                     'rfe': RFE(make_pipeline(LogisticRegression()), n_features_to_select=n_features),
                     'pca': PCA(n_components=n_features)
                    }
        pipeline.append(('feature_selector', selectors.get(selector_to_use)))
            
    pipeline.append(('classifier', model))
    return Pipeline(pipeline)
```

#### Logistic Regression

##### helper function

```python
def perform_lr(X, y, penalty=None, refit='acc', **kwargs):
    lr_pipe = make_pipeline(LogisticRegression(penalty=penalty, solver='saga'), **kwargs)
    print(f"Logistic Regression-{penalty}", "-"*30)
    if penalty is None:
        lr_pipe.fit(X, y)
        return lr_pipe
    else:
        return grid_search_fit(lr_pipe, X, y, is_pipeline=True, refit=refit)
```

```python
def perform_svc(X, y, **kwargs):
    svm = SVC()
    pipe_svm = make_pipeline(svm, **kwargs)
    params = {'kernel': ['linear', 
                         'poly', 
                         'rbf', 
                         'sigmoid'
                        ],
             'C': [0.01, 0.1, 1, 10, 100]}
    return grid_search_fit(pipe_svm, X, y, params=params, is_pipeline=True)
```

#### Custom Ensemble Methods

##### helper functions

```python
def ensemble_predictions(models, X):
    preds = []
    for model in models:
        preds.append(model.predict(X))
    return mode(preds, axis=0).mode


def ensemble_multi_score(models, X, y, runs=10, progress=False, output='all', additional_print="", ret_data=False):
    total_acc = 0
    for i in range(runs):
        y_pred = ensemble_predictions(models, X)
        ps_output = 'acc' if progress else None
        acc, conf_mat, clf_rep = score(y, y_pred, output=ps_output, ret_data=True)
        total_acc += acc
    
    avg_acc = total_acc / runs
    acc_print = f"For {runs} runs: Avg {additional_print} Accuracy={acc}"
    if output == 'all':
        print(acc_print)
        print("Confusion Matrix:", f"Num FP={conf_mat[0, 1]}\n", conf_mat)
        print("Classification Report:\n", clf_rep)
    elif output == 'acc':
        print(acc_print)
    if ret_data:
        return acc, conf_mat, clf_rep
    
    
def master_ensembler(ensemble_methods, X, y, runs=10, progress=False, output='all', additional_print="", ret_data=False, **kwargs):
    total_acc = 0
    for i in range(runs):
        preds = []
        for ensemble in ensemble_methods:
            preds.append(ensemble_predictions(ensemble, X))
        y_pred = mode(preds, axis=0).mode[0]
        ps_output = 'acc' if progress else None
        acc, conf_mat, clf_rep = score(y, y_pred, output=ps_output, ret_data=True)
        total_acc += acc
        
    avg_acc = total_acc / runs
    acc_print = f"For {runs} runs: Avg {additional_print} Accuracy={acc}"
    if output == 'all':
        print(acc_print)
        print("Confusion Matrix:", f"Num FP={conf_mat[0, 1]}\n", conf_mat)
        print("Classification Report:\n", clf_rep)
    elif output == 'acc':
        print(acc_print)
    if ret_data:
        return acc, conf_mat, clf_rep
        

def stacked_X(models, X):
    preds = []
    for model in models:
        preds.append(model.predict(X))
    return np.column_stack(preds)
    
```
